```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("12. Seaborn_Batsman.csv")

df['StrikeRate'] = df['RunsScored'] / df['BallsFaced'] * 100
df['NotOut'] = df['Dismissal'].apply(lambda x: x == 'not out')

avg_runs_year = df.groupby('Year')['RunsScored'].mean()
avg_runs_year.plot(kind='line', marker='o', title='Average Runs Per Year')
```

```python
plt.xlabel('Year')
plt.ylabel('Average Runs')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['Dismissal'].value_counts().plot(kind='bar', title='Dismissal Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_runs_opposition = df.groupby('Opposition')['RunsScored'].mean().sort_values()
avg_runs_opposition.plot(kind='barh', title='Average Runs by Opposition')
plt.xlabel('Average Runs')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='BattingPosition', y='RunsScored')
plt.title('Runs by Batting Position')
plt.show()
```

```python
total_runs = df['RunsScored'].sum()
total_innings = len(df)
not_outs = df['NotOut'].sum()
batting_average = total_runs / (total_innings - not_outs)
strike_rate = total_runs / df['BallsFaced'].sum() * 100
fifties = len(df[df['RunsScored'] >= 50])
hundreds = len(df[df['RunsScored'] >= 100])

print("🏏 Player Career Summary:")
print(f"Total Runs: {total_runs}")
print(f"Innings: {total_innings}, Not Outs: {not_outs}")
print(f"Batting Average: {batting_average:.2f}")
print(f"Strike Rate: {strike_rate:.2f}")
print(f"50s: {fifties}, 100s: {hundreds}")
```
